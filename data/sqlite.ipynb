```python
# ===========================================
# 1) Upload CSV file to Colab
# ===========================================
from google.colab import files
import pandas as pd
import sqlite3

print("Please upload your CSV file:")
uploaded = files.upload()

# Get CSV filename
csv_filename = list(uploaded.keys())[0]

# ===========================================
# 2) Load CSV into DataFrame
# ===========================================
df = pd.read_csv(csv_filename)

print("\n----- CSV Loaded Successfully -----")
print(df.head())

# ===========================================
# 3) Create SQLite Database and Insert CSV
# ===========================================
db_name = "mydatabase.db"
table_name = "mytable"

conn = sqlite3.connect(db_name)

df.to_sql(table_name, conn, if_exists="replace", index=False)

print(f"\n----- Data Inserted into SQLite DB: {db_name}, Table: {table_name} -----")

# ===========================================
# 4) Run SQL Query
# ===========================================
query = f"SELECT * FROM {table_name} LIMIT 10;"
result = pd.read_sql(query, conn)

print("\n----- SQL Query Result -----")
print(result)

conn.close()
```
